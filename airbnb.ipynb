```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
app_df = pd.read_csv('googleplaystore.csv')
app_df = app_df.dropna()
app_df.head()
```

```python
del app_df['Type']
```

```python
del app_df['Content Rating']
```

```python
del app_df['Last Updated']
```

```python
del app_df['Current Ver']
```

```python
del app_df['Android Ver']
```

```python
app_df.corr()
```

```python
def process_size(size):
    if 'M' == size[-1]:
        return float(size[:-1])
    else:
        return 0.0
```

```python
app_df['Size'] = app_df['Size'].apply(lambda x : process_size(x))
```

```python
app_df.head()
```

```python
app_df['Installs'].unique()
```

```python
app_df['Installs_int'] = pd.Categorical(app_df['Installs']).codes
app_df.head(20)
```

```python
app_df['Installs_int'] = pd.Categorical(app_df['Installs'], ['1+', '5+', '10+']).codes
app_df.head(20)
```

```python
install_types = list(app_df['Installs'].unique())
install_types.sort(key=lambda x : int(x[:-1].replace(',', '')))
install_types
```

```python
app_df['Installs_int'] = pd.Categorical(app_df['Installs'], install_types).codes
app_df.head(20)
```

```python
app_df['Category_int'] = pd.Categorical(app_df['Category']).codes
app_df.head(20)
```

```python
app_df['Genres_int'] = pd.Categorical(app_df['Genres']).codes
app_df.head(10)
```

```python
app_df['Reviews'] = app_df['Reviews'].astype(int)
```

```python
app_df.corr()
```

```python
app_df['Price'].unique()
```

```python
def process_price(price):
    if '$' == price[0]:
        return float(price[1:])
    else:
        return float(price)
```

```python
app_df['Price'] = app_df['Price'].apply(lambda x : process_price(x))
```

```python
app_df.corr()
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.feature_extraction import stop_words
```

```python
def pre_process(text):
    
    # lowercase
    text=text.lower()
    
    #remove tags
    text=re.sub("<!--?.*?-->","",text)
    
    # remove special characters and digits
    text=re.sub("(\\d|\\W)+"," ",text)
    
    return str(text)
```

```python
app_df['app_name_processed'] = app_df['App'].apply(lambda x : pre_process(x))
app_df.head()
```

```python
app_df['app_name_processed']
```

```python
app_df['app_name_processed'].sample(20)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.75, stop_words=stop_words.ENGLISH_STOP_WORDS, use_idf=True)
```

```python
names_tfidf_vectors = tfidf_vectorizer.fit_transform(app_df['app_name_processed'])
```

```python
names_tfidf_vectors.toarray()
```

```python
import numpy as np
data = np.asarray(names_tfidf_vectors.toarray())
```

```python
extra_data = app_df[['Size', 'Price']].values
```

```python
extra_data
```

```python
data = np.append(data, extra_data, axis=1)
```

```python
data
```

```python
labels = app_df['Category_int'].values
```

```python
labels
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_validation, y_train, y_validation = train_test_split(data, labels, test_size=0.2)
```

```python
clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_validation, y_validation)
```

```python
probabilitys = clf.predict_proba(x_validation)
predictions = clf.predict(x_validation)
```

```python
predictions[0]
```

```python
probabilitys[0]
```

```python
max_probabilitys = [max(probability) for probability in probabilitys]
```

```python
max_probabilitys
```

```python
def get_confidence_accuracy(threshold, max_probabilitys, predictions, y_validation):
    total = 0
    correct = 0
    size = len(predictions)
    for i in range(size):
        if max_probabilitys[i] > threshold:
            total = total + 1
            
            if predictions[i] == y_validation[i]:
                correct = correct + 1
    
    print('Remain data = %d' % total)
    print('Correct data = %d' % correct)
    return correct / total
```

```python
get_confidence_accuracy(0.2, max_probabilitys, predictions, y_validation)
```

```python
get_confidence_accuracy(0.25, max_probabilitys, predictions, y_validation)
```

```python
get_confidence_accuracy(0.3, max_probabilitys, predictions, y_validation)
```
