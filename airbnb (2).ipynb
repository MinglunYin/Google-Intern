```python
df['room_type']= df['room_type'].astype('category')
df['room_type_val'] = df['room_type'].cat.codes
```

```python
train_column = ['price', 'latitude', 'longitude', 'availability_365', 'room_type_val']
```

```python
from sklearn.preprocessing import MinMaxScaler

for column in train_column:
    scaler = MinMaxScaler()
    df[column + '_scaled'] = scaler.fit_transform(df[[column]].values)
```

```python
scaler = MinMaxScaler()
df['reviews_per_month_scaled'] = scaler.fit_transform(df[['reviews_per_month']].values)
```

```python
df.head(50)
```

```python
validation_rows = df.sample(4000)
```

```python
validation_rows.head(50)
```

```python
train_rows = df.drop(validation_rows.index)
```

```python
train_rows.head(50)
```

```python
scaled_train_columns = [column + '_scaled' for column in train_column]
```

```python
train_data = train_rows[scaled_train_columns].values
validation_data = validation_rows[scaled_train_columns].values
```

```python
train_labels = train_rows['reviews_per_month_scaled']
validation_labels = validation_rows['reviews_per_month_scaled']
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(train_data, train_labels)
```

```python
clf.score(validation_data, validation_labels)
```
